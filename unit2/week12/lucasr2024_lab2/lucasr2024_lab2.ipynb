```python
#Lucas Ross Nov 16 2022

#import the libraries here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load titanic.csv . HINT : first download the dataset in colab
titanic = pd.read_csv("titanic.csv")
titanic
```

```python
# Subset the titanic dataset to include first class passengers who embarked in Southampton
south = titanic[titanic["Embarked"] == "S"] #apply conditional to subset the df
first = titanic[titanic["Pclass"] == 1]
firstSouth = pd.merge(south, first) #only take the colums that are the same
firstSouth
```

```python

# Subset the titanic dataset to include either second or third class passenger
second = titanic[titanic["Pclass"] == 2]
third = titanic[titanic["Pclass"] == 3]
secondThird = pd.concat([second, third]) #concatenate the two dfs
secondThird
```

```python
#print the first 10 rows of firstSouth to read the dataset
print(firstSouth.head(10))
```

```python
#print the first 10 rows of secondthird to read the dataset
print(secondThird.head(10))
```

```python
# Create a bar chart for the first class passengers who embarked in Southampton grouped by sex

#split firstSouth by sex
male = firstSouth[firstSouth["Sex"] == "male"]
female = firstSouth[firstSouth["Sex"] == "female"] #length/amount of rows in df --> len(df)

#make bar chart
plt.bar("male", len(male), color="blue")
plt.bar("female", len(female), color="pink")

plt.title("First Class Southampton Passengers by Sex")
plt.ylabel("Number of Passengers")
plt.xlabel("Sex")

plt.show()
```

```python
# Create a bar chart for the second and third class passengers grouped by survival status

#split secondThird by survival status
alive = secondThird[secondThird["Survived"] == 1]
dead = secondThird[secondThird["Survived"] == 0] #length/amount of rows in df --> len(df)

#make bar chart
plt.bar("alive", len(alive), color="green")
plt.bar("dead", len(dead), color="blue")

plt.title("Second and Third Class Passengers by Survival Status")
plt.ylabel("Number of Passengers")
plt.xlabel("Survival Status")

plt.show()
```
