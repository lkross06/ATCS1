```python
#Lucas Ross 7 Dec 2022

#import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#array of random values in a 10x10 matrix
data = np.random.randint(low=1, high=100, size=(10, 10))
print(data)
```

```python
'''
HEATMAP
- higher val --> lighter color ("hotter")
- lower val --> darker color ("colder")

- 0-9 on axes = each indp var, concentration gradient = y axis
- at (0,0), (1,1), (2,2), etc. the indp var is compared to itself

- good for densities and numeric values
'''
sns.heatmap(data)
```

```python
'''
less sensitive heatmap
 >= vmax --> vmax color
 <= vmin --> vmin color
'''
sns.heatmap(data, vmin=25, vmax=75, annot=True) #annot = True --> puts values of each matrix cell on graph
```
