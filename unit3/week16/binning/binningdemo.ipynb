```python
#Lucas Ross 3 Feb 2023

#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("housing.csv")
df.head(10)
```

```python
'''
numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
'''

bins = np.linspace(df.median_house_value.min(), df.median_house_value.max(), 4) #give it a range (min-max) and # bins (4)
print(bins) #gives intervals between ranges of values
```

```python
#pd.cut makes a categorical column based on the bins and the median_house_value column
#by default it does not include the lowest value
df["bins"] = pd.cut(df.median_house_value, bins, labels = ["low", "medium", "high"], include_lowest=True)
df
```
